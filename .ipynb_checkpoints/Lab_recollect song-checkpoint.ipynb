```python
def get_audio_features_and_titles(user_id, playlist_id):

    results = sp.user_playlist_tracks(user_id, playlist_id)
    tracks = results ['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])

    audio_features = []
    song_names =[]
    song_artist=[]
    for x in range(len(tracks)):
        audio_features.extend(sp.audio_features(tracks=tracks[x]['track']['id']))
        song_names.append(tracks[x]['track']['name'])
        song_artist.append(tracks[x]["track"]["artists"][0]["name"])
    df_name = pd.DataFrame(audio_features)
    df_songname = pd.DataFrame(song_names)
    df_artist = pd.DataFrame(song_artist)

    df_full = pd.concat([df_name, df_songname, df_artist], axis=1, join='inner')

    return df_full
```

```python
K = range(2, 20)
inertia = []

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=1234)
    kmeans.fit(pool_scaled)
    inertia.append(kmeans.inertia_)

inertia
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.show()
```

```python
K = range(2, 20)

silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(pool_scaled)
    silhouette.append(silhouette_score(pool_scaled, kmeans.predict(pool_scaled)))


plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.show()
```

```python
def recommend_song(full_frame):
    # get song id
    song_name = input('Choose a song: ')
    results = sp.search(q=f'track:{song_name}', limit=1)
    track_id = results['tracks']['items'][0]['id']
    
    # get song features with the obtained id
    audio_features = sp.audio_features(track_id)
    
    # create dataframe
    df_ = pd.DataFrame(audio_features)
    new_features = df_[pool_scaled.columns]
    
    # scale features
    scaled_x = scaler.transform(new_features)
    # predict cluster
    cluster = kmeans.predict(scaled_x)
    
    # filter dataset to predicted cluster
    filtered_df = full_frame[full_frame['cluster'] == cluster[0]][pool_scaled.columns]
    
    
    # get closest song from filtered dataset
    closest, _ = pairwise_distances_argmin_min(scaled_x, filtered_df)
    # return it in a readable way
    print('\n [RECOMMENDED SONG]')
    return ' - '.join([full_frame.loc[closest]['song_name'].values[0], full_frame.loc[closest]['artist'].values[0]])
recommend_song(full_frame)
```

```python

def song_rec (titles, artist_names):
    titles= input('insert song')
    artist_names = input('insert artist')
    searchQuery = titles + ' ' + artist_names
    results = sp.search(q=searchQuery)

    full_frame 
    ###first passing by hot 100
    
    if titles in song_names and artist_names in artist:
        print('\n [your song is a hot 100 one! heres the RECOMMENDED SONG]')
        return df.sample()           
    elif bool(results):
    #getting the id trough name and artist    
        try:
            track_id = results['tracks']['items'][0]['id']
        except:
            print('\n The song you typed does not exist or contains an error, please type again')
    # get song features with the obtained id
        audio_features = sp.audio_features(track_id)
    
    # create dataframe
        df_ = pd.DataFrame(audio_features)
        new_features = df_[pool_scaled.columns]
    
    # scale features
        scaled_x = scaler.transform(new_features)
    # predict cluster
        cluster = kmeans.predict(scaled_x)
    
    # filter dataset to predicted cluster
        filtered_df = full_frame[full_frame['cluster'] == cluster[0]][pool_scaled.columns]
    
    
    # get closest song from filtered dataset
        closest, _ = pairwise_distances_argmin_min(scaled_x, filtered_df)
    # return it in a readable way
        print('\n [your song is not in the top 100, here is a simmilar song]')
        return ' - '.join([full_frame.loc[closest]['song_name'].values[0], full_frame.loc[closest]['artist'].values[0]])
    
    else:
        print(' song doesnt exist')
        return 'song doesnt exist'

song_rec(titles, artist_names)
```
