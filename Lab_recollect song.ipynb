```python
def get_audio_features_and_titles(user_id, playlist_id):

    results = sp.user_playlist_tracks(user_id, playlist_id)
    tracks = results ['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])

    audio_features = []
    song_names =[]
    song_artist=[]
    for x in range(len(tracks)):
        audio_features.extend(sp.audio_features(tracks=tracks[x]['track']['id']))
        song_names.append(tracks[x]['track']['name'])
        song_artist.append(tracks[x]["track"]["artists"][0]["name"])
    df_name = pd.DataFrame(audio_features)
    df_songname = pd.DataFrame(song_names)
    df_artist = pd.DataFrame(song_artist)

    df_full = pd.concat([df_name, df_songname, df_artist], axis=1, join='inner')

    return df_full
```
